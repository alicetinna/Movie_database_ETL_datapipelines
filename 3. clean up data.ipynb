```python
import os
import pandas as pd
```

```python
cwd=os.getcwd() # dirname of absolute path of file
Base_DIR=os.path.dirname(cwd)
Data_DIR=os.path.join(Base_DIR ,'data')
cache_DIR=os.path.join(Base_DIR ,'cache')
working_file=os.path.join(cache_DIR , "movies-box_office_dataset.csv")
os.makedirs(cache_DIR , exist_ok=True)
output_file=os.path.join(cache_DIR , "movies-box_office_dataset-cleaned.csv")
```

```python
df = pd.read_csv(working_file)
df["Rank"]=-1
```

```python
df.head(n=1200)
```

```python
df['Domestic %']=df['%']
df['Foreign %']=df["%.1"]
df.head(2)
```

```python
df.drop(['%','%.1'], axis=1 , inplace=True)
```

```python
df.head(4)
```

```python
#df['Worldwide']=df['Worldwide'].replace("$","").replace(",","")
```

```python
to_clean_cols=['Worldwide','Domestic','Foreign']

def currency_string_to_int(current_val):
    currency_val=current_val.replace("$","").replace(",","")
    try:
        currency_val = int(currency_val)
    except:
        currency_val = 0
    return currency_val

def clean_col(row):
    #print(row)
    for col in to_clean_cols:
        current_val=row[col]
        row[col]=currency_string_to_int(current_val)
    return row

df_cleaned = df.apply(clean_col , axis=1)
print(df_cleaned.head())
```

```python
df_cleaned.dtypes
```

```python
df_cleaned.sort_values(by=['Worldwide'] , inplace=True , ascending=False)
df_cleaned.head()
```

```python
df_cleaned.reset_index(inplace=True , drop=True)
df_cleaned.head()
```

```python
df_cleaned['Rank']=df_cleaned.index + 1
df_cleaned.head()
```

```python
df_cleaned['Domestic %'] =df_cleaned['Domestic']/df_cleaned['Worldwide']
df_cleaned['Foreign %'] =df_cleaned['Foreign']/df_cleaned['Worldwide']
df_cleaned.head()
```

```python
df_cleaned.to_csv(output_file , index=False)
```
